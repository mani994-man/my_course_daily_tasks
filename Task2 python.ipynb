```python
def area_of_circle(radius):
    return 3.1416 * radius * radius

print(area_of_circle(7))
```

```python
def largest(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

print(largest(100, 25, 15))
```

```python
def string_length(s):
    count = 0
    for _ in s:
        count += 1
    return count

print(string_length("Manikanta"))
```

```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

print(is_prime(78))
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

print(fibonacci(5))
```

```python
def reverse_string(s):
    if s == "":
        return s
    return reverse_string(s[1:]) + s[0]

print(reverse_string("Spare me great lord"))
```
